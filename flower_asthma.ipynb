```python
# 날짜에 맞는 파일이 다 들어왔는지 확인
# 5년 간 연도별로 참나무의 4,5,6월치 데이터 = 182개
cnt_16, cnt_17, cnt_18, cnt_19, cnt_20 = 0,0,0,0,0

files = os.listdir('./꽃가루농도/')
for file in files:
    if file[-14:-10] == '2016':
        cnt_16 += 1
    elif file[-14:-10] == '2017':
        cnt_17 += 1
    elif file[-14:-10] == '2018':
        cnt_18 += 1
    elif file[-14:-10] == '2019':
        cnt_19 += 1
    elif file[-14:-10] == '2020':
        cnt_20 += 1
    else:
        print(file)
#     print("{}-{}-{}".format(file[-14:-10], file[-10:-8], file[-8:-6]))

print(cnt_16, cnt_17, cnt_18, cnt_19, cnt_20)
```

```python
# 참나무

start = time.time()
exp_year = [2016,2017,2018,2019,2020]
exp_month = [4,5,6]
exp_time = ['06', '18']

column_names = ['plantType', 'areaNo','datetime','todayPred','tomPred','aftertomPred', '=']
df_all = pd.DataFrame(columns = column_names)

# dt_parser = lambda x: datetime.strptime(x, "%Y%m%d%H")

# 날짜별, 시간별 파일명 생성

# 연도
for year in exp_year:
    # 월
    for month in exp_month:
        # 일
        for date in range(1, 32):
            # 관측 시간
            for time in exp_time:
                filename = './꽃가루농도/FCT_IDX_D06_' + '%0d'%year + '%02d'%month + '%02d'%date + time +'.csv'
                if date ==1 and time=='06':
                    print(f'Merging {filename} and now there are {len(df_all)} rows...')
                try:
                    file = pd.read_csv(filename, delimiter='#', engine='python')
                    file.columns = ['plantType', 'areaNo','datetime','todayPred','tomPred','aftertomPred', '=']
                    file['datetime'] = pd.to_datetime(file['datetime'], format="%Y%m%d%H")
                    df_all = pd.concat([df_all, file], axis=0)
                except Exception as e:
                    print(e, 'for ', '%0d-'%year + '%02d-'%month + '%02d-'%date + time)
        
print("============ {} min =============".format((time.time() - start)/60.0))
print(df_all.shape)
df_all.head()
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import calendar
from math import sin, cos, sqrt, atan2, radians
from folium import FeatureGroup, LayerControl, Map, Marker
from folium.plugins import HeatMap
import matplotlib.dates as mdates
import matplotlib as mpl
```

```python
# 지도에 그리려면 null값 포함 안해야 함
# 지표가 0 보다 크고 지역코드가 존재하는 경우
cleaned = merged[(merged['todayPred'] > 0.0) & (merged['areaNo'] > 0) & (merged['경도(초/100)'] > 0.0)]
```

```python
min_lat, max_lat = cleaned.loc[cleaned['경도(초/100)'] > 0.0, '경도(초/100)'].min(), cleaned.loc[cleaned['경도(초/100)'] > 0.0, '경도(초/100)'].max(); print(min_lat, max_lat)
min_long, max_long = cleaned.loc[cleaned['위도(초/100)'] > 0.0, '위도(초/100)'].min(), cleaned.loc[cleaned['위도(초/100)'] > 0.0, '위도(초/100)'].max(); print(min_long, max_long)

pickup_map = Map(location=[35.5838699,127.0565831], zoom_start = 6.4,)

hm_wide = HeatMap( list(zip(cleaned['위도(초/100)'].values, cleaned['경도(초/100)'].values, cleaned['todayPred'])),
                     min_opacity=0.9,
                     radius=5, blur=15,
                     max_zoom=1 
                 )
pickup_map.add_child(hm_wide)

pickup_map
```

### 1단계만 존재하는 경우 = 광역시, 특별시, 도

```python
# 2,3단계가 null -> 광역시, 특별시, 도단위

first_loc = merged[(merged['1단계'].notnull()) &(merged['2단계'].isnull()) & (merged['3단계'].isnull())]

# 행정구역/ 연도별 수치
yearly_loc_sum = first_loc.groupby(['year','1단계'])['todayPred'].sum().reset_index()

yearly_loc_sum.to_csv("./행정구역1단계_연도별_당일지표_합계.csv")

# plt.figure(figsize=(10,8))
# sns.barplot(x='')
```

#### 광역시, 도를 크게 수도권, 호남, 영남, 호서, 강원으로 나눠볼까요

```python
first_loc['1단계'].unique()
```

```python
first_loc.loc[first_loc['1단계'].isin(['서울특별시','인천광역시','경기도']), '지방구분'] = '수도권'
first_loc.loc[first_loc['1단계'].isin(['부산광역시','경상북도','경상남도', '울산광역시','대구광역시']), '지방구분'] = '영남권'
first_loc.loc[first_loc['1단계'].isin(['강원도']), '지방구분'] = '강원도'
first_loc.loc[first_loc['1단계'].isin(['전라북도','전라남도','광주광역시']), '지방구분'] = '호남권'
first_loc.loc[first_loc['1단계'].isin(['충청북도','충청남도','대전광역시','세종특별자치시']), '지방구분'] = '호서권'
first_loc.loc[first_loc['1단계'].isin(['전라북도','전라남도','광주광역시']), '지방구분'] = '호남권'
first_loc.loc[first_loc['1단계'].isin(['제주도']), '지방구분'] = '제주도'

first_loc.sample(4)
```

```python
first_loc['권역평균위험도'] = first_loc.groupby('지방구분')['todayPred'].mean()
first_loc.head()
```
